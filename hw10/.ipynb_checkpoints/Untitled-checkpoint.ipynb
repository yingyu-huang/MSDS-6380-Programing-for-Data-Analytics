```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#from datetime import datetime
#import matplotlib.dates as mdates

# Read-in the data

data=pd.read_excel("Superstore.xls")
data.head()
```

```python
dollarAmt = data['Sales']*data['Discount']
profit=data['Profit']
a = plt.figure()
plt.scatter(dollarAmt,profit, marker = 'o')
plt.xlabel('Discount Amount')
plt.ylabel('Profit')
a.savefig("a.pdf")
```
