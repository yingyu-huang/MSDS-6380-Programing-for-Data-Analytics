```python
# Homework 10 by Team 17
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#from datetime import datetime
#import matplotlib.dates as mdates

# Read-in the data
df = pd.ExcelFile("Superstore.xls")
nf = np.array(df)

# Question 1
# Compute the discount for each transaction
sheetX = df.parse('Orders') 
discount = sheetX['Discount']
sales = sheetX['Sales']
dollarAmt = sales*discount
# Create a scatter plot of profit and discount amount
fig1 = plt.figure()
profit = sheetX['Profit']
plt.xlabel('Discount Amount')
plt.ylabel('Profit')
plt.scatter(dollarAmt,profit, marker = 'o')
fig1.savefig("Question1.pdf")
```

```python
# Question 2
# Group the transactions according to the profitability
def pltcolor(lst):
    cols=[]
    for l in lst:
        if l > 0:           # profitable
            cols.append('orange')
        elif l < 0:         # unprofitable
            cols.append('blue')
    return cols
cols=pltcolor(profit)
plt.scatter(x=dollarAmt,y=profit,c=cols)
plt.xlim(0, 3500)
plt.ylim(-14000, 14000)
plt.xlabel('Discount Amount')
plt.ylabel('Profit')  
plt.savefig("Question2.pdf")
```

```python
# Question 3
# Display sales by the order date
df = pd.read_excel('Superstore.xls')
# Set index to order date and group by months
df['month'] = pd.to_datetime(df['Order Date']).dt.to_period('M')
months = df['month'].sort_values()
dates = np.array(months)
month_1 = dates[0]
month_2 = dates[-1]
index = pd.period_range(start = month_1, end = month_2)
df.groupby('month')['Sales'].sum().reindex(index).plot()
plt.xlabel('Month of Order date')
plt.ylabel('Sales')  
loc = [0,50000,100000,150000,200000,250000,300000,350000,400000,450000,500000]
labels = ["0k","50k","100k","150k","200k","250k","300k","350k","400k","450k","500k"]
plt.yticks(loc, labels)
plt.savefig("Question3.pdf")
```

```python
# Question 4
# Group all the Sales
df["different"] = df.apply(lambda x:'Negative Profit' if x.Profit>0 else 'Positive Profit' ,axis=1) 
df.groupby(['month','different'])['Sales'].sum().unstack().plot(color=('r','g'))  # Plot with colors
plt.yticks(loc, labels)
plt.xlabel('Month of Order date')
plt.ylabel('Sales')
plt.savefig("Question4.pdf")
```

```python
# Question 5
# Plot a bar graph to show percentages of category across regions
Superstore_counts = pd.crosstab(df['Region'], df['Product Category'])
Superstore_pcts = Superstore_counts.div(Superstore_counts.sum(1), axis=0)
print(Superstore_pcts)
Superstore_pcts.plot.bar()
plt.savefig("Question5.pdf")
```

```python
# Question 6
# Visualize the relationship between these three variables: profit, sales, and shipping cost
x = df['Sales']
y = df['Profit']
s = df['Shipping Cost']
plt.scatter(x,y,s=s ,facecolors = 'none',edgecolors = 'b')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.savefig("Question6.pdf")
```

```python
# Question 7
# Display order quantity by shipping mode via boxplots
sns.catplot(x ='Ship Mode', y='Order Quantity', kind = 'box',data = df)
plt.savefig("Question7.pdf")
```
