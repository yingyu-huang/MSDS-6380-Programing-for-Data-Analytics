```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Read-in data
data=pd.read_excel("C:/Users/yingy/Desktop/52 python/hw8/Ecoli Data.xlsx")
```

```python
data['year']=data["Date Collected"].dt.year
```

```python
year_gropued=data.groupby(data['year'])['E.coli'].mean()
year_gropued
```

```python
#Plot a bar graph in which x-axis is years (2013-2016) and y-axis is E.coli amount. 
year_gropued.plot('bar')

plt.show()
```

```python
data['month']=data["Date Collected"].dt.month
month_gropued=data.groupby(data['month'])['E.coli'].mean()

plt.show(month_gropued.plot('bar'))
```

```python
ecoli_corr=pd.Series({'E.coli-Specific Conductance':data['E.coli'].corr(data['Specific Conductance']),
                     'E.coli-Dissolved Oxygen':data['E.coli'].corr(data['Dissolved Oxygen']),
                     'E.coli-PH':data['E.coli'].corr(data['PH']),
                     'E.coli-Residue':data['E.coli'].corr(data['Residue']),
                     'E.coli-Ammonia Nitrogen':data['E.coli'].corr(data['Ammonia Nitrogen']),
                     'E.coli-Nitrate Nitrogen':data['E.coli'].corr(data['Nitrate Nitrogen']),
                     'E.coli-Nitrogen':data['E.coli'].corr(data['Nitrogen']),
                     'E.coli-Phosphorous':data['E.coli'].corr(data['Phosphorous']),
                     'E.coli-Chloride':data['E.coli'].corr(data['Chloride']),
                     'E.coli-Sulfate':data['E.coli'].corr(data['Sulfate'])})
print(ecoli_corr)

plt.show(ecoli_corr.plot('bar'))
```
